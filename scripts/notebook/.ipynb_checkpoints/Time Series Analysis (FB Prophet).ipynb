```python
directoryPath = "../../exports"
project_name = "angular"
task = "BUG"
file = "{0}/{1}/{2}_dataset_{3}.csv"

df = pd.read_csv(file.format(directoryPath, project_name, project_name, task))
df = df[[c.DATE, c.NT_CC]]
df.columns = ['ds','y']
# df['cap'] = df['y'].max()
# df['floor'] = 0
```

```python
df.describe()
```

```python
m = Prophet(interval_width=0.90, growth='linear')
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=36, freq='m')
# future['cap'] = df['y'].max()
# future['floor'] = 0
forecast = m.predict(future)
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```

```python
from fbprophet.diagnostics import cross_validation
from fbprophet.diagnostics import performance_metrics

cross_validation_results = cross_validation(m, initial= '1000 days', horizon='365 days')
print(cross_validation_results)

```

```python
performance_metrics_results = performance_metrics(cross_validation_results)
performance_metrics_results.head()
```

```python
print(performance_metrics_results['mape'].median())
print(performance_metrics_results['mdape'].median())
```
