```python
import pandas as pd
from pandas import datetime
from pandas import DataFrame
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.ar_model import AR
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
def parser(x):
    return datetime.strptime(x.strip(), '%Y-%m-%d')

df = pd.read_csv("../../output/data_analysis_major_versions_with_dates.csv", index_col=[' Release Date'], parse_dates=[' Release Date'], date_parser=parser)
E_Line_Effort = df[' E_Line']
```

```python
E_Line_Effort.head()
```

```python
E_Line_Effort.plot()
```

```python
plot_acf(E_Line_Effort)
```

```python
## Stationarity
# mean, variance, covariance is constant over periods.
# for Linux, this does not appear to be the case
# need to convert this to stationary
# e.g. integrated of order 1, denoted by d (for diff), one of the parameter of ARIMA model
E_Line_Effort_Diff = E_Line_Effort.diff(periods=1)
E_Line_Effort_Diff = E_Line_Effort_Diff[1:]
E_Line_Effort_Diff.head()
```

```python
plot_acf(E_Line_Effort_Diff)
```

```python
E_Line_Effort_Diff.plot()
```

```python
## Auto Regressive AR Model
X = E_Line_Effort.values
train = X[0:55] # 55 observations in train data
test = X[55:] # 19 observations in test data
test.size
model_ar = AR(train)
model_ar_fit = model_ar.fit(disp=2)
predictions = model_ar_fit.predict(start=55, end=74)
```

```python
plt.plot(test)
plt.plot(predictions, color='red')
```

```python
## ARIMA Model
# p, d, q:  
# p = periods taken for AR model
# d = integrated order, difference
# q = periods in moving average model
model_arima = ARIMA(train, order=(1,1,1))
model_arima_fit = model_arima.fit(disp=2)
print(model_arima_fit.summary())
```

```python
predictions = model_arima_fit.forecast(steps=19)[0]
predictions
```

```python
model_arima_fit.plot_predict(dynamic=False)
```

```python
# plot residual errors
residuals = DataFrame(model_arima_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

```python
from pmdarima.arima import auto_arima
from plotly.offline import plot_mpl
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(E_Line_Effort, model='additive', period=7)
fig = result.plot()
plot_mpl(fig)
```

```python
stepwise_model = auto_arima(E_Line_Effort, start_p=1, start_q=1,
                           max_p=3, max_q=3, m=12,
                           start_P=0, seasonal=False,
                           d=1, D=1, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
print(stepwise_model.aic())
```

```python
X = E_Line_Effort
train = X[0:55] # 55 observations in train data
test = X[55:] # 19 observations in test data
```

```python
stepwise_model.fit(train)
```

```python
future_forecast = stepwise_model.predict(n_periods=19)
print(future_forecast)
```

```python
future_forecast = pd.DataFrame(future_forecast,index = test.index,columns=['Prediction'])
pd.concat([test,future_forecast],axis=1).plot()
```

```python
pd.concat([E_Line_Effort,future_forecast],axis=1).plot()
```
