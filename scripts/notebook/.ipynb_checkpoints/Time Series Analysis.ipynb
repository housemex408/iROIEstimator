```python
import pandas as pd
from pandas import datetime
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
def parser(x):
    return datetime.strptime(x.strip(), '%Y-%m-%d')

df = pd.read_csv("../../output/data_analysis_major_versions_with_dates.csv", index_col=[' Release Date'], parse_dates=[' Release Date'], date_parser=parser)
E_Line_Effort = df[' E_Line']
```

```python
E_Line_Effort.head()
```

```python
E_Line_Effort.plot()
```

```python
plot_acf(E_Line_Effort)
```

```python
# Stationarity
# mean, variance, covariance is constant over periods.
# for Linux, this does not appear to be the case
# need to convert this to stationary
# e.g. integrated of order 1, denoted by d (for diff), one of the parameter of ARIMA 
E_Line_Effort_Diff = E_Line_Effort.diff(periods=1)
E_Line_Effort_Diff = E_Line_Effort_Diff[1:]
E_Line_Effort_Diff.head()
```

```python
plot_acf(E_Line_Effort_Diff)
```

```python
E_Line_Effort_Diff.plot()
```
