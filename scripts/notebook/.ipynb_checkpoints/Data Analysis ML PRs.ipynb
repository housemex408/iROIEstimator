```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import os as os
sys.path.append('../../')
import Constants as c
import Utilities as utils
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import model_selection
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import KFold
%matplotlib inline
```

```python
def add_NT(df, task_type):
    nt = df.copy()
    nt.where(df[c.TASK] == task_type, inplace = True)
    nt.dropna(axis = 0, how = 'all', inplace = True)
    nt = nt[["N_T", c.VERSION]]

    if nt.isna().values.any():
        nt.fillna(0, inplace=True)
        
    df = df.join(nt.set_index(c.VERSION), on=c.VERSION, rsuffix='_'+task_type)
    return df
```

```python
directoryPath = "../../exports"
project_name = "angular"
task = "BUG"
file = "{0}/{1}/{2}_PRs.csv"

df = pd.read_csv(file.format(directoryPath, project_name, project_name))
df = add_NT(df, "DOCS")
df = add_NT(df, "REFACTOR")
df = add_NT(df, "TESTING")
df = add_NT(df, "FEATURE")
df = add_NT(df, "UPGRADE")
df = add_NT(df, "RELEASE")
df = add_NT(df, "SUPPORT")
df = add_NT(df, "OTHER")

nt_columns = ["N_T_DOCS", "N_T_REFACTOR", "N_T_TESTING", "N_T_FEATURE", "N_T_UPGRADE", "N_T_RELEASE", "N_T_SUPPORT", "N_T_OTHER"]
df["NO"] = df[nt_columns].sum(axis=1)

df.where(df[c.TASK] == task, inplace = True)
df.dropna(axis = 0, how = 'all', inplace = True)

if df.isna().values.any():
    df.fillna(0, inplace=True)
```

```python
# df = utils.isRegularVersion(df)
# get_NT(df, "DOCS")
# get_NT(df, "REFACTOR")
# get_NT(df, "TESTING")
# get_NT(df, "FEATURE")
# get_NT(df, "UPGRADE")
# get_NT(df, "RELEASE")
# get_NT(df, "SUPPORT")
# get_NT(df, "OTHER")

df[c.DATE] = pd.to_datetime(df[c.DATE])
df['diff_hours'] = df[c.DATE].diff(-1) * (-1) / np.timedelta64(1, 'h')
df['year'] = df[c.DATE].dt.year
df['month'] = df[c.DATE].dt.month
df['day'] = df[c.DATE].dt.day
df['quarter'] = df[c.DATE].dt.quarter
df.drop(df.columns[0], axis=1, inplace=True)
# df.drop(c.DATE, axis=1, inplace=True)
# df.drop(c.PROJECT, axis=1, inplace=True)
# df.drop(c.VERSION, axis=1, inplace=True)
# df.drop("f0_", axis=1, inplace=True)

#Capping the outlier rows with Percentiles
# upper_lim = df[c.MODULE].quantile(.95)
# lower_lim = df[c.MODULE].quantile(.05)
# df.loc[(df[c.MODULE] > upper_lim),c.MODULE] = upper_lim
# df.loc[(df[c.MODULE] < lower_lim),c.MODULE] = lower_lim

# upper_lim = df[c.LINE].quantile(.95)
# lower_lim = df[c.LINE].quantile(.05)
# df.loc[(df[c.LINE] > upper_lim),c.LINE] = upper_lim
# df.loc[(df[c.LINE] < lower_lim),c.LINE] = lower_lim


percent_25 = df[c.MODULE].quantile(.25)
percent_50 = df[c.MODULE].quantile(.50)
percent_75 = df[c.MODULE].quantile(.75)

df['Module_Bin'] = pd.cut(df[c.MODULE], bins=[df[c.MODULE].min(),percent_25,percent_50,percent_75, df[c.MODULE].max()], labels=[0, 1,2,3], duplicates='drop')
df.head()

percent_25 = df[c.LINE].quantile(.25)
percent_50 = df[c.LINE].quantile(.50)
percent_75 = df[c.LINE].quantile(.75)

df['Line_Bin'] = pd.cut(df[c.LINE], bins=[df[c.LINE].min(),percent_25,percent_50,percent_75, df[c.LINE].max()], labels=[0,1,2,3], duplicates='drop')
df.head()

conditions = [
    df[c.VERSION].str.contains('-rc'),
    df[c.VERSION].str.contains('-beta'),
    df[c.VERSION].str.contains('-alpha')
]

choices = [2, 3, 4]

df['Release'] = np.select(conditions, choices, default=1)

if df.isna().values.any():
    df.fillna(0, inplace=True)
```

```python
X = df[["N_T", "NO", c.T_MODULE, c.T_LINE, "T_Commits", "Contributions", "Contributors", "Comments", "Commenters", 'Release', 'Line_Bin', 'quarter', 'Release', 'diff_hours', 'year', 'month', 'day']]
# X = df[[c.NT, c.NO, 'Line_Bin', c.T_CONTRIBUTORS ]]
# X = df[["N_T", "T_Commits", "Contributions", "Contributors", "Comments", "Commenters"]]
Y = df[c.LINE]

splits = 10
num_records = len(X)

if num_records <= splits:
    splits = num_records

# model = DecisionTreeRegressor(random_state=0)
model = RandomForestRegressor(n_estimators=300, n_jobs=-1, random_state=0)
model.fit(X, Y)

kfold = model_selection.KFold(n_splits=splits)
predictions = cross_val_predict(model, X, Y, cv=kfold)
# Interpreting Restuls:  https://www.youtube.com/watch?v=U7D1h5bbpcs  
```

```python
data = {}
data[c.OBSERVED] = Y.round(2)
data[c.PREDICTED] = predictions.round(2)
data[c.DIFFERENCE] = abs(Y - predictions).round(2)
data[c.PERCENT_ERROR] = (abs(Y - predictions)/Y).round(2)
results = pd.DataFrame(data)
results['Version'] = np.arange(results.__len__())
print(round(utils.calculate_PRED(0.25, results, c.PERCENT_ERROR), 2))
print(round(utils.calculate_PRED(0.50, results, c.PERCENT_ERROR), 2))
# results
```
