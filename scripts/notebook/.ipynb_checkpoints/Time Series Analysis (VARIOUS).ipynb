```python
import pandas as pd
from pandas import datetime
from pandas import DataFrame
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.ar_model import AR
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from pmdarima.arima import auto_arima
from plotly.offline import plot_mpl
from statsmodels.tsa.seasonal import seasonal_decompose
from math import sqrt
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing, Holt
sys.path.append('../../')
import Constants as c
import Utilities as utils
from fbprophet import Prophet
```

```python
variable = ' E_Module'
releaseDate = ' Release Date'

df = pd.read_csv("../../output/data_analysis_major_versions_with_dates.csv", usecols=[variable, releaseDate])
df[releaseDate] = pd.to_datetime(df[releaseDate])
df = df.set_index(releaseDate)
df.index = df.index.strftime('%Y-%m-%d') 
df.head()
```

```python
train = df[0:55] # 55 observations in train data
test = df[55:] # 19 observations in test data
```

```python
result = seasonal_decompose(train, model='additive', period=12)
fig = result.plot()
plot_mpl(fig)
```

```python
arima_model = auto_arima(train, m=12)
print(arima_model.aic())
```

```python
forecast = arima_model.predict(n_periods=len(test))
data = {variable: forecast, releaseDate: test.index} 
forecast = pd.DataFrame(data) 
forecast = forecast.set_index(releaseDate)
```

```python
#plot the predictions for validation set
plt.figure(figsize=(22, 10))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test', color='r')
plt.plot(forecast.index, forecast, label='SARIMA', color='k')
plt.legend(loc='best', fontsize='xx-large')
plt.show()
```

```python
#calculate rmse
mse = mean_squared_error(test,forecast)
print(mse)

rms = sqrt(mse)
print(rms)

mae = mean_absolute_error(test,forecast)
print(mae)
```

```python
simple_series = pd.Series(df[variable], df.index)


simple_smoothing_model = SimpleExpSmoothing(simple_series).fit()
simple_forecast = simple_smoothing_model.forecast(12).rename(r'$\alpha=%s$'%simple_smoothing_model.model.params['smoothing_level'])
# plot
simple_forecast.plot(marker='o', color='green', legend=True)
simple_smoothing_model.fittedvalues.plot(marker='o', color='green')

plt.show()
```

```python
holt_model = Holt(simple_series, damped=True).fit(smoothing_level=0.8, smoothing_slope=0.2)
holt_forecast = holt_model.forecast(12).rename("Additive damped trend")

holt_model.fittedvalues.plot(marker="o", color='green')
holt_forecast.plot(color='green', marker="o", legend=True)

plt.show()
```

```python
holt_winters_add = ExponentialSmoothing(simple_series, seasonal_periods=12, trend='add', seasonal='add').fit(use_boxcox=True)
holt_winters_mul = ExponentialSmoothing(simple_series, seasonal_periods=12, trend='add', seasonal='mul').fit(use_boxcox=True)

holt_winters_add.fittedvalues.plot(style='--', color='red')
holt_winters_mul.fittedvalues.plot(style='--', color='green')

holt_winters_add.forecast(12).plot(style='--', marker='o', color='red', legend=True)
holt_winters_mul.forecast(12).plot(style='--', marker='o', color='green', legend=True)

plt.show()
```

```python
holt_winters_add = ExponentialSmoothing(train, seasonal_periods=12, trend='add', seasonal='add').fit(use_boxcox=True)
predictions = holt_winters_add.forecast(19).rename('Forecast')

data = {variable: predictions, releaseDate: test.index} 
forecast = pd.DataFrame(data) 
forecast = forecast.set_index(releaseDate)

plt.figure(figsize=(22, 10))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test', color='r')
plt.plot(forecast.index, forecast, label='Holt-Winters', color='k')
plt.legend(loc='best', fontsize='xx-large')
plt.show()
```

```python
mse = mean_squared_error(test,forecast)
print(mse)

rms = sqrt(mse)
print(rms)

mae = mean_absolute_error(test,forecast)
print(mae)
```
